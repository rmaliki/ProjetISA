# MODEL FOR 363 lines Gas

```python
#Splitting training to train and validation set
X_train, X_valid = train_test_split(train, test_size=0.4, random_state=10)
X_train1, X_valid1 = train_test_split(train1, test_size=0.4, random_state=10)
```

```python
test.API=(test.API).astype('int64')
test1.API=(test1.API).astype('int64')
test.Zone=(test.Zone).astype('int64')
test1.Zone=(test1.Zone).astype('int64')
```

```python
Y_train =(X_train.GasCum360)
Y_train1=(X_train1.GasCum360)
#
X_train = X_train.drop(["GasCum360"], axis=1)
X_train = X_train.drop(["OilCum360"], axis=1)
X_train1 = X_train1.drop(["GasCum360"], axis=1)
X_train1 = X_train1.drop(["OilCum360"], axis=1)
#
dtrain = xgboost.DMatrix(X_train, Y_train)
dtrain1 = xgboost.DMatrix(X_train1, Y_train1)
#
Y_valid = (X_valid.GasCum360)
Y_valid1 = (X_valid1.GasCum360)
#
X_valid = X_valid.drop(["GasCum360"], axis=1)
X_valid = X_valid.drop(["OilCum360"], axis=1)
dvalid = xgboost.DMatrix(X_valid, Y_valid)
X_valid1 = X_valid1.drop(["GasCum360"], axis=1)
X_valid1 = X_valid1.drop(["OilCum360"], axis=1)
dvalid1 = xgboost.DMatrix(X_valid1, Y_valid1)
#
X_test = test
X_test1 = test1
#
dtest = xgboost.DMatrix(X_test)
dtest1 = xgboost.DMatrix(X_test1)
```

```python
def rmspe_xg(yhat, y):
    y =y.get_label()
    yhat = yhat
    return "rmspe", rmspe(y, yhat)
def rmspe(y, yhat):
    return np.sqrt(np.mean(((y - yhat) / y) ** 2))
```

```python
###
watchlist = [(dtrain, "train"), (dvalid, "eval")]
print("Training model...")
params = {"objective": "reg:linear",
          "booster": "gbtree",
          "eta": 0.01,
          "max_depth": 10,
          "subsample": 0.8,
          "colsample_bytree": 0.7,
          "silent": 1,
          "seed": 1301}
num_boost_round = 1000
model = xgboost.train(params, dtrain, num_boost_round, evals=watchlist,
                      early_stopping_rounds=100, feval=rmspe_xg, verbose_eval=True)
pickle.dump(model, open("model.dat", "wb"))
```

```python
#Validation
yhat = model.predict(xgboost.DMatrix(X_valid))
error = rmspe(Y_valid,yhat)
print("RMSPE: {:.6f}".format(error))
```

```python
###
watchlist1 = [(dtrain1, "train"), (dvalid1, "eval")]
print("Training model...")
params1 = {"objective": "reg:linear",
          "booster": "gbtree",
          "eta": 0.01,
          "max_depth": 10,
          "subsample": 0.8,
          "colsample_bytree": 0.4,
          "silent": 1,
          "seed": 1301}
num_boost_round1 = 1000
model1 = xgboost.train(params1, dtrain1, num_boost_round1, evals=watchlist1,
                      early_stopping_rounds=100, feval=rmspe_xg, verbose_eval=True)
pickle.dump(model1, open("model1.dat", "wb"))
```

```python
#Validation
yhat = model1.predict(xgboost.DMatrix(X_valid1))
error = rmspe(Y_valid1,yhat)
print("RMSPE: {:.6f}".format(error))
```

```python
#Prediction on test set of splitted data grp1
predictions = model.predict(dtest)
```
