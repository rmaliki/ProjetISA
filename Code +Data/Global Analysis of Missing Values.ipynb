```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import outlier_removal as out
from sklearn import linear_model
df=pd.read_csv('TrainSample.csv',sep=';' )
from  numpy import nan
```

## Analysis by Lines

```python
missing=df.isnull().as_matrix()
count_na_lines=[[i,np.sum(missing[i,:])] for i in range(len(missing))]
t=pd.DataFrame(count_na_lines)
t.columns = ['line','nb of nas per row']
t=t.groupby('nb of nas per row')['line'].count()
t=t.to_frame()
t.columns = ['amount of rows']
t
```

Above we can see that we have two big groups of rows (records).We define:

Group1: less than 5    missing values. 363 members

Group2: more than 18 missing values. 97 members

### seperate the two groups

```python
l1=[i for i in range(len(df)) if count_na_lines[i][1]<=6 ]
l2=[i for i in range(len(df)) if count_na_lines[i][1]>6 ]
group1=df.iloc[l1]
group2=df.iloc[l2]
group2
```

### Analyze group 2
 We would like to know whether the missing values in group2 are in the same columns or each row has the missing values in different columns. A look into the data frame of group2 (above) confirms, that for group2, there is a set of 18 columns that is completely blank. We count the number of nans for each column of group2:

```python
missing2=group2.isnull().as_matrix()
count_na_columns=[[i,np.sum(missing2[:,i])] for i in range(len(missing2[0]))] # substract data frame index
s=pd.DataFrame(count_na_columns)
s.columns = ['column','nb of nans in column']
s=s.groupby('nb of nans in column')['column'].count()
s=s.to_frame()
s.columns = ['amount of columns']
s['amount of columns'][0]=s['amount of columns'][0]-2  # data frame index given by python and API do not count 
s
```

We can confirm, that there is a set of 18 columns, which have all the 97 rows of group 2 missing.

We count for each column of the whole dataset, how many missing values it has.

```python
count_na_columns=[[i,np.sum(missing[:,i])] for i in range(len(missing2[0]))] # substract data frame index
s=pd.DataFrame(count_na_columns)
s.columns = ['column','nb of nans in column']
s=s.groupby('nb of nans in column')['column'].count()
s=s.to_frame()
s.columns = ['amount of columns']
s['amount of columns'][0]=s['amount of columns'][0]-2  # data frame index given by python and API do not count 
s
```
